### 목표
- 학생 수학 성적 예측하는 회귀 모델을 만들어보자!
- keras를 이용하여 딥러닝 신경망을 구성해보자

```python
# 데이터 불러오기
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("./data/student-mat.csv", delimiter=";")
```

```python
data
```

```python
# 컬럼명만 확인
data.columns

# 답 데이터 : G3 (수학 성적) 
# 특성 데이터 : studytime(공부할 시간) 특성 1개 학습 진행
```

```python
# 문제와 답 분리
X = data['studytime'] # 공부한 시간 1차원
y = data['G3'] #G3 1차원 - 회귀

# Train, test 분리
# 7:3분리, 난수 : 3
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=3)
```

```python
# 크기확인
print("훈련용",X_train.shape,y_train.shape)
print("테스트용",X_test.shape,y_test.shape)

```

### 신경망 모델 만들기
- 1. 신경망 구조 설계 : 뼈대 구축, 층 쌓아주기
- 2. 신경망 학습/ 평가 방법 설정
- 3. 학습 및 학습 결과 시각화
- 4. 모델 평가 및 예측

```python
# 1. 신경망 구조 설계

# 뼈대 구축 도구
from tensorflow.keras import Sequential

# 층 쌓기 
from tensorflow.keras.layers import Dense, Activation

# Dense : 층 내용 정의 부분

# Activation : 활성화 함수

# 뼈대 구축
model = Sequential()

# 층 쌓기 -> 입력층, 중간(은닉)층, 출력층
# 유닛 == 뉴런
# 입력층 + 중간층(뉴런 개수가 4개로 구성)
# Dense(유닛 개수, 입력 데이터 크기 설정)
model.add(Dense(units=4, input_dim = 1))
model.add(Activation('sigmoid')) # 활성화 함수

# 출력 층
# 성적 예측 -> 회귀
# units의 개수는 항상 1
# 활성화 함수 : 'linear' 앞에 층과 같은 모양으로 생략이 가능하다. => 항등 함수

model.add(Dense(units=1))

# 요약
model.summary()
```

```python
# 2. 신경망 학습/평가 방법
# compile()
# 모델의 설계를 pc가 이해할 수 있도록 번역하는 과정

# 오차 계산 공식, 최적화 함수 , 평가지표
model.compile(loss="mse", #손실함수 (비용함수):mse(평균제곱오차)=> 회귀일때)
                optimizer = "sgd" #최적화 함수 : 경사하강법 종류 설정 => 확률적 경사하강법
              ) # metrics는 생략 평가지표는 mse로 활용
```

```python
# 학습
# h 변수에 학습 현황 저장
h = model.fit(X_train,y_train, epochs=200)
```

```python
# 모델 평가
model.evaluate(X_test, y_test)
```

```python
# 모델 예측

model.predict(X_test)
```
